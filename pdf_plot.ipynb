## Calcolo funzioni di struttura Lagrangiane di ordine p: $$ S_p^L(\tau) = <(v_x(t+\tau)-(v_x(t))^p> $$

## Kurtosis o Flatness: $$ K^L(\tau) = \frac{S_4^L(\tau)}{[S_2^L(\tau)]^2} $$

```python
struct_real_more = np.loadtxt(f"data/struct_function_30000_part.dat")
struct_real_less = np.loadtxt(f"data/struct_function_10000_part.dat")
run = 9
number = 9999
#struct_gen_less = np.loadtxt(f"data/struct_function_10000_part_gen_wgangp_{run}_{number}_.dat")
struct_gen_more = np.loadtxt(f"data/struct_function_30000_part_gen_wgangp_{run}_{number}_5punti.dat")
semidisp = (10.273698864467972-(-9.970374739869616))/2.

import matplotlib.pyplot as plt
import matplotlib as mpl
mpl.rcParams.update({'font.size': 22})



plt.figure(figsize=(10,7))
plt.title("Longitudinal Structure Functions")
plt.yscale('log')
plt.xscale('log')
for ii in range(1,4):
    plt.plot(struct_real_more[:,0],struct_real_more[:,ii],label="(R) p= "+str((ii)*2),marker='.',lw=0.4,ms=8, color='C'+str(ii-1))
for ii in range(1,4):
    plt.plot(struct_gen_more[:,0],struct_gen_more[:,ii],label="(G) p= "+str((ii)*2),marker='^',lw=0.4,ms=8, color='C'+str(ii-1))
plt.legend(ncol=2)
plt.xlabel("$\\tau$")
plt.ylabel('$S_p^L(\\tau)$')
plt.tight_layout()
plt.savefig(f"plots/struct_func_comp_gen_{run}_{number}",fmt="png",dpi=100)
plt.show()
```

```python
struct_real_log = np.log(struct_real_more)
struct_gen_log = np.log(struct_gen_more)
dl_struct_real = np.zeros(shape=struct_real_more.shape)
dl_struct_gen = np.zeros(shape=struct_gen_more.shape)
for ii in range(1,4):
    for jj in range(struct_real_more.shape[0]-1):
        dl_struct_real[jj,ii] = (np.log(struct_real_more[jj+1,ii]) - np.log(struct_real_more[jj,ii]))/(np.log(struct_real_more[jj+1,0])-np.log(struct_real_more[jj,0]))
        dl_struct_gen[jj,ii] = (np.log(struct_gen_more[jj+1,ii]*semidisp**(ii*2)) - np.log(struct_gen_more[jj,ii]*semidisp**(ii*2)))/(np.log(struct_gen_more[jj+1,0])-np.log(struct_gen_more[jj,0]))

plt.figure(figsize=(10,7))
plt.title("Longitudinal Structure Functions Local slope")
#plt.yscale('log')
plt.xscale('log')
for ii in range(1,4):
    plt.plot(struct_real_more[:,0],dl_struct_real[:,ii],label="(R) p= "+str((ii)*2),marker='.',lw=0.4,ms=8, color='C'+str(ii-1))
for ii in range(1,4):
    plt.plot(struct_gen_more[:,0],dl_struct_gen[:,ii],label="(G) p= "+str((ii)*2),marker='^',lw=0.4,ms=8, color='C'+str(ii-1))
plt.legend(ncol=2)
plt.xlabel("$\\tau$")
plt.ylabel('$S_p^L(\\tau)$')
plt.tight_layout()
plt.savefig(f"plots/struct_func_loc_slope_comp_gen_{run}_{number}",fmt="png",dpi=100)
plt.show()
```

```python
plt.figure(figsize=(10,7))
plt.title("Flatness S4/S2^2 (K for Kurtosis)\nand S6/S2^3 (H for Hypertailedness)")
# plt.yscale('log')
plt.xscale('log')
plt.yscale('log')

plt.plot(struct_real_more[:,0],struct_real_more[:,2]/struct_real_more[:,1]**2.,marker='.', label="(K)Real", lw=0.4,ms=8,color='C0')
plt.plot(struct_gen_less[:,0],struct_gen_less[:,2]/struct_gen_less[:,1]**2.,marker='^', label="(K)Gen",lw=0.4,ms=8,color='C1')
#plt.plot(struct_real_more[:,0],[3.]*len(struct_real_more[:,0]),label="Gaussian")

plt.plot(struct_real_more[:,0],struct_real_more[:,3]/struct_real_more[:,1]**3.,marker='.', label="(H)Real", lw=0.4,ms=8,color='C2')
plt.plot(struct_gen_less[:,0],struct_gen_less[:,3]/struct_gen_less[:,1]**3.,marker='^', label="(H)Gen",lw=0.4,ms=8,color='C3')

#plt.ylabel("$K(\\tau)$")
plt.xlabel("$\\tau$")
plt.legend(ncol=2)
plt.grid(which='minor', alpha=0.2)
plt.grid(which='major', alpha=0.9)
plt.tight_layout()
plt.savefig(f"plots/struct_func_flatness_comp_gen_{run}_{number}",fmt="png",dpi=100)
plt.show()
```
