IMPORTO DATABASE H5, LO CONVERTO IN NPY E LO SALVO

```python
import h5py
import numpy as np
path = '/scratch/bonaccorso/LagrangianTracers/velocities.h5'
f = h5py.File(path,'r')
dbh5 = f['vel3d']
db = np.ndarray(shape=(327680,2000,3))
i=0
for traj in dbh5:
    db[i,:,:] = traj
    i += 1
#np.save("velocities",db1)
```

VISUALIZZO

```python
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
import numpy as np
%matplotlib inline
n = int(np.random.uniform(0,len(db[:,0,0])))
print(n)
plt.plot(db[n,:,:])
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.plot3D(db[n,:,0],db[n,:,1],db[n,:,2])
```
